```python
# Dropping duplicates, keeping first value, should be the latest timepoint per mouse 
filtered_df = filtered_df.drop_duplicates(subset="Mouse ID", keep='first') 
filtered_df.count()
```

```python
# First create a subset of the combined_data dataframe where Drug Regimen is
# in [Capomulin, Ramicane, Infubinol, and Ceftamin].
# Create a list of the four drugs to examine
four_drug_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Slice the original combined_data dataframe using the list of four drugs
four_drugs_review_df = combined_results_df[combined_results_df['Drug Regimen'].isin(four_drug_list)]
four_drugs_review_df.head()
```

```python
# Then groupby and use the .agg method, chain a round method to output 3 digits after the zero
four_last_tumor_df= four_drugs_review_df.groupby(['Drug Regimen','Mouse ID']).agg(\

# Get the last value of Tumor Volume (mm3) column using a lambda function selecting the element in -1 position
Last_tumor_size=('Tumor Volume (mm3)',lambda x: x.iloc[-1])).round(3)
four_last_tumor_df
```

```python
# Review dataframe with Drug Regimen columns by using unstack and move Last_tumor_size to index using stack
four_last_tumor_unstack_df = four_last_tumor_df.stack(level=0).unstack(level=0)
four_last_tumor_unstack_df.head()
```

```python
#Calculate the IQR and quantitatively determine if there are any potential outliers.

# Set counter to 0 
counter = 0

# Quartile calculations per drug
for drug in four_drug_list:
    quartiles = four_last_tumor_unstack_df[drug].quantile([.25,.5,.75]).round(2)
    lowerq = quartiles[0.25].round(2)
    upperq = quartiles[0.75].round(2)
    iqr = round(upperq-lowerq,2)
    lower_bound = round(lowerq - (1.5*iqr),2)
    upper_bound = round(upperq + (1.5*iqr),2)
    
    # first time through the for loop, if true then print dividing line
    if counter == 0:
        print(f"---------------------------------------------------")
   
    print(f"{drug} IQR data is:")
    print(f"The lower quartile of {drug} is: {lowerq}")
    print(f"The upper quartile of {drug} is: {upperq}")
    print(f"The interquartile range of {drug} is: {iqr}")
    print(f"The the median of {drug} is: {quartiles[0.5]} ")
    print(f"Values below {lower_bound} for {drug} could be outliers.")
    print(f"Values above {upper_bound} for {drug} could be outliers.")
    print(f"---------------------------------------------------")
    counter += 1
```

```python
# Create lists containing each column in the table and remove NaN values

# Create empty list
boxplot_list = []
# Review the list of the four drugs names using them to select the same column names
# and append to the boxplot_list
for drug in four_drug_list:
    boxplot_list.append(list(four_last_tumor_unstack_df[drug].dropna()))
```

```python
# Plot the lists using a boxplot utilizing the same y-axis for all plots
fig1, ax = plt.subplots(figsize=(9,7))
ax.set_title('Final measured Tumor volume by Drug Regimen')
ax.set_xlabel('Drug Regimen')
ax.set_ylabel('Tumor Volume (mm3)')
ax.boxplot(boxplot_list,notch=0,sym='gD')
plt.xticks([1,2,3,4],four_drug_list)

# Save the figure
plt.savefig("output/Box-and-Whisker-Plots.png")

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

# Pull values for x and y values 
mouse_weight = Capomulin_df.groupby(Capomulin_df["Mouse ID"])["Weight (g)"].mean()
tumor_volume = Capomulin_df.groupby(Capomulin_df["Mouse ID"])["Tumor Volume (mm3)"].mean()
y=mouse_weight
x=tumor_volume
```

```python
# Perform a linear regression on year versus violent crime rate
slope, intercept, r, p, std_err = st.linregress(mouse_weight, tumor_volume)
```

```python
# Create equation of line to calculate predicted violent crime rate
fit = slope * mouse_weight + intercept
```

```python
## Plot the linear model on top of scatter plot - OUTPUT ##
plt.scatter(mouse_weight,tumor_volume)
reqress_values=x*slope+intercept
line_eq="y = "+str(round(slope,2))+"x + "+str(round(intercept,2))
plt.plot(mouse_weight,fit,"--",color='r')
plt.xticks(mouse_weight, rotation=0)
plt.annotate(line_eq,(12,40),fontsize=15,color='red')
plt.title("Capomulin Regimen: Mouse Weight vs. Average Tumor Volume")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mmg3)")

# Save chart to file 
plt.savefig("output/scatter-plot-chart-linear-model.png") 

plt.show()
```
